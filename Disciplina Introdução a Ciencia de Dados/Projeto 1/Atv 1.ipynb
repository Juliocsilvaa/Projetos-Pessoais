# **Preparação dos Dados**

```python
banco_de_dados['Data'] = pd.to_datetime(banco_de_dados["Data"],dayfirst=True)
banco_de_dados = banco_de_dados.set_index(["Data","NivelConsistencia"])

banco_de_dados
```

```python
#transformando todos os dados que estao como "object" em "float"
for i in banco_de_dados.columns:
  if(banco_de_dados[i].dtype=='object'):
    banco_de_dados[i]=banco_de_dados[i].str.replace(',','.').astype(float)
banco_de_dados
```

```python
resultados=[]
for (data, nivel), linha in banco_de_dados_filtrado.iterrows():
  ano, mes = data.year, data.month
  _,periodo= calendar.monthrange(ano,mes)
  dias = pd.date_range(start=data, periods= periodo, freq='D')
  for data, (coluna, valor) in zip(dias, linha.items()):
    resultados.append({'Data': data, 'NivelConsistencia': nivel,'Valor': valor})
banco_de_dados_filtrado_arrumado = pd.DataFrame(resultados)
banco_de_dados_filtrado_arrumado
```

```python
#contando os valores nulos
banco_de_dados_filtrado_arrumado['Valor'].isnull().sum()
```

# **Analise**

```python
#Criando um df com as informaçoes
analise = banco_de_dados_filtrado_arrumado.copy()

# Removendo valores ausentes e duplicadas, mantendo a linha com o maior nível de consistência
analise=analise.dropna(subset=['Valor'])
analise=analise.sort_values(by='NivelConsistencia', ascending=False).drop_duplicates(subset='Data', keep='first')
analise=analise.sort_values(by='Data', ascending=True).reset_index(drop=True)

#extraindo as caracteristicas uteis
analise['ano']=analise['Data'].dt.year
analise['mes']=analise['Data'].dt.month
analise['dia']=analise['Data'].dt.day
analise['dias_com_chuva'] = (analise['Valor'] > 0).astype(int)
analise
```

```python
#verificando duplicadas
analise['Data'].value_counts()
```

```python
#Criando um df com todos os dias para a analise de falhas
todas_datas =  pd.date_range(start=analise['Data'].min(), end=analise['Data'].max())
todas_datas_df = pd.DataFrame(todas_datas, columns= ['data'])
todas_datas_df['data'] = pd.to_datetime(todas_datas_df['data'])
todas_datas_df['ano']=todas_datas_df['data'].dt.year
todas_datas_df['mes']=todas_datas_df['data'].dt.month
todas_datas_df['dia']=todas_datas_df['data'].dt.day
todas_datas_df['falhas'] = (~todas_datas_df['data'].isin(analise['Data'])).astype(int)
todas_datas_df
```

## **Analise Anual**

### Analise Padrão

```python
#Agrupando por ano
analise_ano_agrupado = analise.groupby('ano')[['Valor','dias_com_chuva']].sum().reset_index()
analise_ano_agrupado
```

```python
#Agrupando por ano
falhas_agrupado_ano = todas_datas_df.groupby('ano')['falhas'].sum().reset_index()
falhas_agrupado_ano
```

```python
analise_ano_agrupado=pd.merge(analise_ano_agrupado,falhas_agrupado_ano , on='ano')
analise_ano_agrupado
```

```python
# Criar o hietograma com hover indicando falhas
fig = px.bar(analise_ano_agrupado, x='ano', y='Valor',
             title='Hietograma Anual da Chuva',
             labels={'Valor': 'Quantidade de Chuva (mm)', 'ano': 'Ano'},
             text='Valor',
             hover_data={'falhas': True,'dias_com_chuva': True, 'ano': False})

# Discretizar os dados (exibir barras separadas por ano)
fig.update_layout(xaxis=dict(tickmode='array', tickvals=analise_ano_agrupado['ano']))
fig.update_traces(texttemplate='%{y:.1f}', hovertemplate='Ano: %{x}<br>Chuva: %{y:.1f} mm<br>Dias com dados ausentes: %{customdata[0]}<br>Dias com Chuva: %{customdata[1]}')

fig.show()
```

### **Analise Boxplot Anual**

```python
analise_boxsplot_anual = analise[['Valor','ano']].merge(falhas_agrupado_ano,on='ano', how="left")
analise_boxsplot_anual
```

```python
fig = px.box(analise_boxsplot_anual, x="ano", y="Valor", title="Distribuição Anual da Chuva",
             labels={"ano": "Ano", "Valor": "Chuva (mm)"},
             template="plotly_white",
             custom_data=["falhas"])  # Adicionando as falhas no hover

# Atualizando o texto do hover para exibir as falhas
fig.update_traces(hovertemplate="<b>Ano:</b> %{x}<br>" +
                                "<b>Chuva (mm):</b> %{y}<br>" +
                                "<b>Falhas:</b> %{customdata[0]}")

fig.update_layout(xaxis_title="Ano", yaxis_title="Precipitação (mm)", showlegend=False)
fig.show()
```

## **Analise Mensal**

### **Analise da media mensal**

```python
analise_mensal_media = analise.groupby('mes').agg(valor_total=("Valor", "sum"),dias_com_chuva_mensal=("dias_com_chuva", "sum"), anos_distintos=("ano", "nunique")).reset_index()
analise_mensal_media['valor_medio'] = analise_mensal_media['valor_total'] / analise_mensal_media['anos_distintos']
analise_mensal_media['dias_com_chuva_mensal_media'] = analise_mensal_media['dias_com_chuva_mensal'] / analise_mensal_media['anos_distintos']
analise_mensal_media
```

```python
falhas_agrupado_mes = todas_datas_df.groupby('mes').agg(valor_total=("falhas", "sum"), anos_distintos=("ano", "nunique")).reset_index()
falhas_agrupado_mes['valor_medio_falhas'] = falhas_agrupado_mes['valor_total']/falhas_agrupado_mes['anos_distintos']
falhas_agrupado_mes
```

```python
analise_mensal_media_agrupado = pd.merge(analise_mensal_media,falhas_agrupado_mes, on='mes')
analise_mensal_media_agrupado
```

```python
# Criar um mapeamento manual dos meses em português
nomes_meses = {1: "Janeiro", 2: "Fevereiro", 3: "Março", 4: "Abril", 5: "Maio", 6: "Junho",7: "Julho", 8: "Agosto", 9: "Setembro", 10: "Outubro", 11: "Novembro", 12: "Dezembro"}

# Substituir os números dos meses pelos nomes no DataFrame
analise_mensal_media_agrupado['nome_mes'] = analise_mensal_media_agrupado['mes'].map(nomes_meses)

# Criar o hietograma com hover indicando falhas
fig = px.bar(analise_mensal_media_agrupado, x='nome_mes', y='valor_medio',
             title='Hietograma Mensal Médio da Chuva',
             labels={'valor_medio': 'Quantidade de Chuva (mm)', 'nome_mes': 'Mês'},
             text='valor_medio',
             hover_data={'valor_medio_falhas': ':.2f','dias_com_chuva_mensal_media': True, 'nome_mes': False})

# Discretizar os dados (exibir barras separadas por ano)
#fig.update_layout(xaxis=dict(tickmode='array', tickvals=analise_mensal_media_agrupado['mes']))
fig.update_layout(xaxis=dict(tickmode='array', tickvals=list(nomes_meses.values())))
fig.update_traces(texttemplate='%{y:.1f}', hovertemplate='Mês: %{x}<br>Média de chuva: %{y:.1f} mm<br>Média de dias com dados ausentes: %{customdata[0]:.2f}<br>Média de dias com chuva: %{customdata[1]:.2f}')

fig.show()
```

### **Analise Boxplot Mensal**

```python
analise_boxsplot_mensal = analise[['Valor','mes','ano','dia']].merge(falhas_agrupado_mes[['valor_medio_falhas','mes']],on='mes', how="left")
analise_boxsplot_mensal['nome_mes'] = analise_boxsplot_mensal['mes'].map(nomes_meses)
analise_boxsplot_mensal
```

```python
# Transformando a coluna 'nome_mes' em categórica com a ordem correta
ordem_meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho","Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]
analise_boxsplot_mensal["nome_mes"] = pd.Categorical(analise_boxsplot_mensal["nome_mes"], categories=ordem_meses, ordered=True)

# Criando o gráfico
fig = px.box(analise_boxsplot_mensal, x="nome_mes", y="Valor", title="Distribuição Mensal da Chuva",labels={"nome_mes": "Mês", "Valor": "Chuva (mm)"},template="plotly_white",custom_data=["ano", "dia", "valor_medio_falhas"])

# Texto do hover
fig.update_traces(hovertemplate="<b>Ano:</b> %{customdata[0]}<br>" + "<b>Dia:</b> %{customdata[1]}<br>" + "<b>Chuva (mm):</b> %{y}<br>" + "<b>Média de Falhas:</b> %{customdata[2]:.2f}"
)

# Ajustando os rótulos do eixo
fig.update_layout(xaxis_title="Mês", yaxis_title="Precipitação (mm)", showlegend=False)
fig.show()
```

### **Analise boxsplot mensal acumulado**

```python
#criando o acumulado
analise_boxsplot_mensal_acumulado = analise_boxsplot_mensal.groupby(['ano','mes'])['Valor'].sum().reset_index()
analise_boxsplot_mensal_acumulado['nome_mes'] = analise_boxsplot_mensal_acumulado['mes'].map(nomes_meses)
analise_boxsplot_mensal_acumulado
```

```python
# Transformando a coluna 'nome_mes' em categórica com a ordem correta
analise_boxsplot_mensal_acumulado["nome_mes"] = pd.Categorical(analise_boxsplot_mensal_acumulado["nome_mes"], categories=ordem_meses, ordered=True)

# Criando o gráfico
fig = px.box(analise_boxsplot_mensal_acumulado, x="nome_mes", y="Valor", title="Distribuição AcumuladoMensal da Chuva",labels={"nome_mes": "Mês", "Valor": "Chuva (mm)"},template="plotly_white",custom_data=["ano"])

# Texto do hover
fig.update_traces(hovertemplate="<b>Ano:</b> %{customdata[0]}<br>" + "<b>Chuva (mm):</b> %{y}<br>")

# Ajustando os rótulos do eixo
fig.update_layout(xaxis_title="Mês", yaxis_title="Precipitação (mm)", showlegend=False)
fig.show()
```

## **Analise Diaria**

### Média de cada dia

```python
analise_diaria_media = analise.groupby('dia').agg(valor_total=("Valor", "sum"),dias_com_chuva_diario=("dias_com_chuva", "sum"), dias_distintos=("dia", "count")).reset_index()
analise_diaria_media['valor_medio'] = analise_diaria_media['valor_total'] / analise_diaria_media['dias_distintos']
analise_diaria_media['%_de_chuva_diario'] = (analise_diaria_media['dias_com_chuva_diario'] / analise_diaria_media['dias_distintos'])*100
analise_diaria_media
```

```python
falhas_agrupado_dia = todas_datas_df.groupby('dia').agg(valor_total=("falhas", "sum"), dias_distintos=("dia", "count")).reset_index()
falhas_agrupado_dia['%_de_falhas'] = (falhas_agrupado_dia['valor_total']/falhas_agrupado_dia['dias_distintos'])*100
falhas_agrupado_dia
```

```python
analise_diaria_media_agrupado = pd.merge(analise_diaria_media,falhas_agrupado_dia, on='dia')
analise_diaria_media_agrupado
```

```python
# Criar o hietograma com hover indicando falhas
fig = px.bar(analise_diaria_media_agrupado, x='dia', y='valor_medio',
             title='Hietograma Diario Médio da Chuva',
             labels={'valor_medio': 'Quantidade de Chuva (mm)', 'dia': 'Dias'},
             text='valor_medio',
             hover_data={'%_de_falhas': ':.2f','%_de_chuva_diario': True, 'dia': False})

# Discretizar os dados (exibir barras separadas por ano)
fig.update_layout(xaxis=dict(tickmode='array', tickvals=analise_diaria_media_agrupado['dia']))
#fig.update_layout(xaxis=dict(tickmode='array', tickvals=list(nomes_meses.values())))
fig.update_traces(texttemplate='%{y:.1f}', hovertemplate='Dia: %{x}<br>Média de chuva: %{y:.1f} mm<br>Porcentagem de ter dado ausente: %{customdata[0]:.2f}%<br>Porcentagem de chuva: %{customdata[1]:.2f}%')

fig.show()
```
